````python
# Dimensiones
dimensiones = df.shape

# Tipos de datos
tipos_datos = df.dtypes


# Valores nulos
valores_nulos = df.isnull().sum()


# Estad√≠sticas descriptivas
estadisticas = df.describe(include='all')


print("Dimensiones:", dimensiones)
print("\nTipos de datos:\n", tipos_datos)
print("\nValores nulos:\n", valores_nulos)
print("\nEstad√≠sticas descriptivas:\n", estadisticas)
````

````python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder

# Definici√≥n de columnas
drop_columns = ["link", "direccion"]
numeric_standard = ["n_habitaciones", "n_ba√±os", "n_estacionamientos"]
numeric_minmax = ["total_superficie_m2", "superficie_construida_m2", "valor_uf"]
categorical_nominal = ["comuna", "quien_vende", "corredor"]

# ColumnTransformer
preprocesador = ColumnTransformer(
    transformers=[
        ("num_std", StandardScaler(), numeric_standard),
        ("num_mm", MinMaxScaler(), numeric_minmax),
        ("cat_nom", OneHotEncoder(handle_unknown="ignore", sparse_output=False), categorical_nominal)

    ],
    remainder="drop"
)

# Aplicar transformaciones
df_procesado = preprocesador.fit_transform(df.drop(columns=drop_columns))

# Nombres de columnas
column_names = preprocesador.get_feature_names_out()

# Crear DataFrame con nombres de columnas
df_procesado = pd.DataFrame(df_procesado, columns=column_names)

# Mostrar tama√±o final
print(df_procesado.shape)
print(df_procesado.head())
````

> **Respuestas aqu√≠**

El dataset se preproces√≥ eliminando las variables irrelevantes (link, direccion), estandarizando las num√©ricas discretas (n_habitaciones, n_ba√±os, n_estacionamientos) con StandardScaler, y normalizando las continuas (superficie_m2, valor_uf) con MinMaxScaler.
Las variables categ√≥ricas (comuna, quien_vende, corredor) se transformaron con OneHotEncoder por ser nominales sin orden.
El resultado fue un df_procesado de dimensi√≥n (1064, 454) con todas las variables listas para el modelado.

### 3. `KMeans` [0.2 Puntos]

En las siguiente secciones usaremos KMeans para encontrar patrones interesantes en los datos.

A continuaci√≥n:

- [ ] Clusterice los datos utilizando KMeans con 5 centros.
- [ ] Guarde los clusters generados en una nueva columna del dataframe `df`.
- [ ] Utilice el c√≥digo provisto para graficar los clusters generados.
- [ ] Utilice `df.groupby("kmeans_cluster").mean().round(2)` para resumir las variables num√©ricas que representan a cada cluster y comente que observa.

Opcional:

- Intente proyectar los centros de los clusters y luego proyectelos (sin puntaje).

````python
# ===========================================================
# 1) Importar librer√≠as
# ===========================================================
from sklearn.cluster import KMeans
import plotly.express as px
from umap import UMAP

# ===========================================================
# 2) Proyecci√≥n UMAP (guardar el modelo para transformar centros)
#    Nota: si ya lo ten√≠as, solo aseg√∫rate de conservar el objeto 'umap_model'
# ===========================================================
umap_model = UMAP(n_neighbors=20, min_dist=0.25, random_state=10)
proj = umap_model.fit_transform(df_procesado)

df.loc[:, "x"] = proj[:, 0]
df.loc[:, "y"] = proj[:, 1]

# ===========================================================
# 3) KMeans con k=5
# ===========================================================
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
df["kmeans_cluster"] = kmeans.fit_predict(df_procesado)

# ===========================================================
# 4) Graficar puntos + proyectar centros al 2D de UMAP
# ===========================================================
fig = px.scatter(
    df,
    x="x",
    y="y",
    color="kmeans_cluster",
    hover_name="comuna",
    hover_data=["valor_uf", "n_habitaciones", "n_ba√±os"],
    title="Clusters con KMeans (k=5) en proyecci√≥n UMAP"
)

# Proyectar centros KMeans al espacio 2D (usar el mismo UMAP)
centers_2d = umap_model.transform(kmeans.cluster_centers_)

fig.add_scatter(
    x=centers_2d[:, 0],
    y=centers_2d[:, 1],
    mode="markers",
    marker=dict(size=14, symbol="x"),
    name="Centros"
)

fig.show()

# ===========================================================
# 5) Resumen num√©rico por cluster (evita columnas no num√©ricas)
# ===========================================================
num_cols = df.select_dtypes(include="number").columns
resumen = df.groupby("kmeans_cluster")[num_cols].mean().round(2)
print(resumen)
````

> Respuestas aqu√≠

Se identificaron 5 clusters de viviendas con caracter√≠sticas diferenciadas: desde propiedades peque√±as y econ√≥micas (clusters 1 y 4) hasta casas de lujo de gran tama√±o y alto valor UF (cluster 3).
El cluster 2 destaca por combinar gran superficie y muchos estacionamientos, mientras que el cluster 0 muestra viviendas amplias pero casi sin estacionamientos.


### 4. Encontrar el mejor n√∫mero de clusters [0.3 Puntos]

- Utilice el c√≥digo visto en clases para encontrar el mejor n√∫mero de clusters usando el m√©todo del codo y el coeficiente de silhouette.
- Seleccione un n√∫mero de clusters y repita el proceso de la subsecci√≥n anterior (graficar los clusters encontrados). Luego, responda:
  - ¬øPara ustedes como analistas de datos, tiene sentido el n√∫mero de clusters encontrado?
  - Al ejecutar `df.groupby("kmeans_cluster").mean().round(2)`, ¬øTienen sentidos los promedios encontrados?
  - ¬øSon √∫tiles los clusters encontrados?

````python
# C√≥digo aqu√≠
# ============================================
# 1) Imports
# ============================================
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import plotly.express as px

# ============================================
# 2) B√∫squeda de k con Codo e √çndice Silhouette
# ============================================
k_min, k_max = 2, 10
inertias, sils, ks = [], [], list(range(k_min, k_max + 1))

for k in ks:
    kmeans_tmp = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels_tmp = kmeans_tmp.fit_predict(df_procesado)
    inertias.append(kmeans_tmp.inertia_)
    sils.append(silhouette_score(df_procesado, labels_tmp))

# --- Plot codo e √≠ndice silhouette ---
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(ks, inertias, marker='o')
plt.title('M√©todo del codo (Inertia)')
plt.xlabel('k'); plt.ylabel('Inertia')

plt.subplot(1,2,2)
plt.plot(ks, sils, marker='o')
plt.title('Coeficiente de silhouette')
plt.xlabel('k'); plt.ylabel('Silhouette')
plt.tight_layout()
plt.show()

# ============================================
# 3) Selecci√≥n autom√°tica de k (m√°x. silhouette)
#    (Si quieres forzar otro k, c√°mbialo manualmente)
# ============================================
best_k = ks[int(np.argmax(sils))]
print(f"Mejor k por silhouette: {best_k:.0f} (score={max(sils):.3f})")

# ============================================
# 4) Entrenar KMeans con k √≥ptimo y graficar
# ============================================
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df["kmeans_cluster"] = kmeans.fit_predict(df_procesado).astype(int)

fig = px.scatter(
    df, x="x", y="y",
    color="kmeans_cluster",
    hover_name="comuna",
    hover_data=["valor_uf","n_habitaciones","n_ba√±os"],
    title=f"Clusters con KMeans (k={best_k}) en proyecci√≥n UMAP"
)

# Si tienes umap_model entrenado, proyecta centros al 2D:
try:
    centers_2d = umap_model.transform(kmeans.cluster_centers_)
    fig.add_scatter(
        x=centers_2d[:,0], y=centers_2d[:,1],
        mode="markers", name="Centros",
        marker=dict(size=14, symbol="x", color="black")
    )
except NameError:
    # Si no existe umap_model, omitimos la proyecci√≥n de centros
    pass

fig.show()

# ============================================
# 5) Resumen por cluster (solo num√©ricas)
# ============================================
num_cols = df.select_dtypes(include="number").columns
resumen = df.groupby("kmeans_cluster")[num_cols].mean().round(2)
print(resumen)
````

> Respuestas aqu√≠
1. ¬øPara ustedes como analistas de datos, tiene sentido el n√∫mero de clusters encontrado?

S√≠, porque el m√©todo del codo muestra una inflexi√≥n en k=2‚Äì3 y el coeficiente de silhouette es m√°ximo en k=2, lo que respalda que dos clusters son adecuados.

2. Al ejecutar df.groupby("kmeans_cluster").mean().round(2), ¬øTienen sentido los promedios encontrados?

S√≠, ya que los promedios diferencian claramente entre un grupo de propiedades grandes, con m√°s habitaciones, ba√±os y mayor valor UF, y otro grupo de propiedades m√°s peque√±as y econ√≥micas.
